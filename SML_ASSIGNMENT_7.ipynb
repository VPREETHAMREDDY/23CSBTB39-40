```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
# Step 1: Read the data from Excel file
#file_path = '/content/xl sheet assignment 7.xlsx' # Replace with your actual file path
data = pd.read_csv('/content/Salary_Data.csv')
# Step 2: Identify the feature(s) and target variable
# Assuming the dataset has columns 'Feature' and 'Target', modify based on your data
features = data[['YearsExperience']] # Change 'Feature' to the actual feature column name
target = data['Salary'] # Change 'Target' to the actual target column name
# Step 3: Plot a graph between features and target
plt.scatter(features, target, color='blue')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Feature vs Target')
plt.show()
# Step 4: Fit a linear regression model
regressor = LinearRegression()
regressor.fit(features, target)
# Plotting the best fit line
predicted_target = regressor.predict(features)
plt.scatter(features, target, color='blue')
plt.plot(features, predicted_target, color='red', linewidth=2)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Best Fit Line: Linear Regression')
plt.show()
# Step 5: Find MSE, MAE for test sizes 20% and 25%
# For 20% test size
X_train_20, X_test_20, y_train_20, y_test_20 = train_test_split(features, target, test_size=0.20,
random_state=42)
regressor_20 = LinearRegression()
regressor_20.fit(X_train_20, y_train_20)
y_pred_20 = regressor_20.predict(X_test_20)
# Calculate MSE and MAE for 20% test size
mse_20 = mean_squared_error(y_test_20, y_pred_20)
mae_20 = mean_absolute_error(y_test_20, y_pred_20)
print(f"Test Size 20% - MSE: {mse_20}, MAE: {mae_20}")
# For 25% test size
X_train_25, X_test_25, y_train_25, y_test_25 = train_test_split(features, target, test_size=0.25,
random_state=42)
regressor_25 = LinearRegression()
regressor_25.fit(X_train_25, y_train_25)
y_pred_25 = regressor_25.predict(X_test_25)
# Calculate MSE and MAE for 25% test size
mse_25 = mean_squared_error(y_test_25, y_pred_25)
mae_25 = mean_absolute_error(y_test_25, y_pred_25)
print(f"Test Size 25% - MSE: {mse_25}, MAE: {mae_25}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
# Read the data
data = pd.read_excel('/content/xl sheet assignment 7.xlsx')
# Step 2: Extract features and target
x = data[['Feature 1','Feature 2','Feature 3','Feature 4' ]].values
y = data['Target (Y)'].values
# Split the data into training and testing sets (80% training, 20% testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Create a Linear Regression model
model = LinearRegression()
# Train the model using the training data
model.fit(x_train, y_train)
# Make predictions on the testing data
y_pred = model.predict(x_test)
# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
# Coefficients of the model (for each independent variable)
print("Coefficients:", model.coef_)
# Intercept of the model
print("Intercept:", model.intercept_)
```
